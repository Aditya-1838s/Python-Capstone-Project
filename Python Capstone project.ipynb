__Task 1: Create dataframes and save as csv__

```python
import numpy as np 
import pandas as pd 
```

```python
# Creating Employee Dataframe
employee_data = { 
    "ID": ["A001", "A002", "A003", "A004", "A005"],
    "Name": ["John Alter", "Alice Luxumberg", "Tom Sabestine", "Nina Adgra", "Amy Johny"],
    "Gender": ["M", "F", "M", "F", "F"],
    "City": ["Paris", "London", "Berlin", "Newyork", "Madrid"],
    "Age": [25, 27, 29, 31, 30]
}
df_employee = pd.DataFrame(employee_data)

# Saving Employee DataFrame as CSV 
df_employee.to_csv("employee.csv", index=False)

# Creating Seniority Level DataFrame
seniority_data = {
    "ID": ["A001", "A002", "A003", "A004", "A005"],
    "Designation Level": [2, 2, 3, 2, 3]
}
df_seniority = pd.DataFrame(seniority_data)

# Saving Seniority Level DataFrame as CSV
df_seniority.to_csv("seniority.csv", index = False)

# Creating Project DataFrame
project_data = {
    "ID": ["A001", "A002", "A003", "A004", "A005", "A002", "A005", "A003", "A001", "A003", "A001", "A004", "A004", "A005"],
    "Project": ["Project 1", "Project 2", "Project 3", "Project 4", "Project 5", "Project 6", "Project 7", "Project 8", "Project 9", "Project 10", "Project 11", "Project 12", "Project 13", "Project 14"],
    "Cost": [1002000, 2000000, 4500000, 5500000, np.nan, 680000, 400000, 350000, np.nan, 300000, 2000000, 1000000, 3000000, 200000],
    "Status": ["Finished", "Ongoing", "Finished", "Ongoing", "Finished", "Failed", "Finished", "Failed", "Ongoing", "Finished", "Failed", "Ongoing", "Finished", "Finished"]
}
df_project = pd.DataFrame(project_data)

# Saving Project DataFrame as CSV
df_project.to_csv("project.csv", index = False)
```

__Task 2: Fill Missing Cost Values with Running Average__

```python
# Load the Project DataFrame
df_project = pd.read_csv("project.csv")

# Loop through each row to check for missing values
for i in range(len(df_project)):
    if pd.isnull(df_project.loc[i, "Cost"]): # Check if the cost is missing 
        df_project.loc[i, "Cost"] = df_project["Cost"][:i].mean() # Replace with running average

# Save the updated DataFrame 
df_project.to_csv("project.csv", index = False)
```

__Task 3: Split Name Column__

```python
# Load the Employee DataFrame
df_employee = pd.read_csv("employee.csv")

# Split 'Name' column into 'First Name' and 'Last Name'
df_employee[['First Name', 'Last Name']] = df_employee['Name'].str.split(' ', n=1, expand = True)

#Drop the old 'Name' column
df_employee.drop(columns = ['Name'], inplace = True)

# Save the updated Employee DataFrame
df_employee.to_csv("employee.csv", index = False)
```

```python
df_employee # Result of above code is below
```

__Task 4: Merge All DataFrames and name it Final__

```python
#Load all DataFrames 
df_seniority = pd.read_csv("seniority.csv")
df_final = df_employee.merge(df_seniority, on = "ID").merge(df_project, on = "ID")

#Save the merged DataFrame
df_final.to_csv("final.csv", index = False)
```

```python
df_final # Result of above code is below
```

__Task 5: Add Bonus Column to Final DataFrame__

```python
# Calculate a 5% bonus only for employees who have finished projects
df_final["Bonus"] = df_final.apply(lambda x: x["Cost"] * 0.05 if x["Status"] == "Finished" else 0, axis=1)

# Save the updated DataFrame
df_final.to_csv("final.csv", index=False)
```

__Task 6: Demote Designation Level and Remove Employees__

```python
import numpy as np 
import pandas as pd
```

```python
# Demote employees whose projects have failed 
df_final.loc[df_final["Status"] == "Failed", "Designation Level"] += 1

# Remove employees whose designation level is above 4
df_final = df_final[df_final["Designation Level"] <= 4]

# Save the updated DataFrame
df_final.to_csv("final.csv", index = False)
df_final
```

__Task 8: Promote Employees Aged > 29__

```python
# Load the Final DataFrame
df_final = pd.read_csv("final.csv")

# Ensure the necessary columns exist
if "Age" in df_final.columns and "Designation Level" in df_final.columns:  
    # Promote employees whose age > 29 (subtract 1 but don't go below 1)
    df_final.loc[(df_final["Age"] > 29) & (df_final["Designation Level"] > 1), "Designation Level"] -= 1

# Save the updated DataFrame
df_final.to_csv("final.csv", index=False)

# Display the updated DataFrame
df_final



```

__Task 9: Compute Total project cost for each Employee__

```python
# Group by employee ID and First Name, summing up the project costs
total_proj_cost = df_final.groupby(["ID", "First Name"])["Cost"].sum().reset_index()

# Rename column for clarity
total_proj_cost.rename(columns={"Cost": "Total Cost"}, inplace=True)

# Save the new DataFrame
total_proj_cost.to_csv("total_proj_cost.csv", index=False)

total_proj_cost

```
